# Plotting a displot
- Combines matplotlib.hist function with seaborn kdeplot()
- KDE Plot represents Kernel Density Estimate
- KDE is used to visualise the probability density of a continuous variable

```python
plt.figure(figsize=(10,50))

for i in range(len(creditcard_df.columns)):
    plt.subplot(17, 1, i + 1)
    sns.histplot(data=creditcard_df[creditcard_df.columns[i]],
                 kde=True,
                #  kde_kws={"color": "b", "lw": 3, "label": "KDE"},
                 color="g")
    plt.title(creditcard_df.columns[i])

plt.tight_layout()
plt.show()
```
